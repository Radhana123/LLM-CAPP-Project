```python
from src.feature_tokenizer import FeatureTokenizer
```

```python
tok = FeatureTokenizer()

parts = [
    {
        "part_id":"P001",
        "material":"Aluminum",
        "features":["Hole","Slot"],
        "tolerance":"0.02mm",
        "batch_size":500
    },
    
    {
        "part_id":"P002",
        "material":"Steel",
        "features":["Thread","Groove"],
        "tolerance":"0.01mm",
        "batch_size":200
    },
    
    {
        "part_id":"P003",
        "material":"Titanium",
        "features":["Pocket"],
        "tolerance":"0.005mm",
        "batch_size":50
    },
]

results = [tok.tokenize_part(p) for p in parts]
df = pd.DataFrame(results)
df
```

```python
plt.figure(figsize=(8, 4))
plt.bar(df["part_id"], df["batch_size"],
        color=["#4f7cff", "#2ecc8a", "#f5a623"])
plt.title("Batch Size per Part", fontsize=14)
plt.xlabel("Part ID")
plt.ylabel("Batch Size")
plt.tight_layout()
plt.show()
```
